```python
from sklearn import svm
import numpy
import matplotlib.pyplot as plt
alg = numpy.linalg
```

```python
#actual data
N = 1000 #sample size
#pos
mean_pos = [1,1]
cov_pos = [[1,0],[0,1]]
data_pos = numpy.random.multivariate_normal(mean_pos, cov_pos, N)
```

```python
#neg
mean_neg = [-1,-1]
cov_neg = [[3,0],[0,3]]
data_neg = numpy.random.multivariate_normal(mean_pos, cov_pos, N)
```

```python
#show our dataset
plt.scatter(data_neg[0,:], data_neg[1,:], c='yellow')
plt.scatter(data_neg[:,0], data_neg[:,1], c='yellow')
plt.scatter(data_pos[0,:], data_pos[1,:], c='blue')
plt.scatter(data_pos[:,0], data_pos[:,1], c='blue')
plt.show()
```

```python
#create our test data
#pos
tmean_pos = [1,1]
tcov_pos = [[1,0],[0,1]]
tdata_pos = numpy.random.multivariate_normal(tmean_pos, tcov_pos, N)
#neg
tmean_neg = [-1,-1]
tcov_neg = [[3,0],[0,3]]
tdata_neg = numpy.random.multivariate_normal(tmean_pos, tcov_pos, N)
```

```python
plt.scatter(tdata_neg[0,:], tdata_neg[1,:], c='yellow')
plt.scatter(tdata_neg[:,0], tdata_neg[:,1], c='yellow')
plt.scatter(tdata_pos[0,:], tdata_pos[1,:], c='blue')
plt.scatter(tdata_pos[:,0], tdata_pos[:,1], c='blue')
plt.show()
```
